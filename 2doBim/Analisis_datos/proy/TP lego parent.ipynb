```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import seaborn as sns; sns.set()

DATASET_PATH = "dataset"
sets = pd.read_csv(DATASET_PATH + "/sets.csv")
themes = pd.read_csv(DATASET_PATH + "/themes.csv")
themes.rename(columns = {'id':'theme_id'}, inplace = True)

sets_themes = sets.merge(themes, on='theme_id', how='left', suffixes=('_sets', '_themes'))

sets_themes.groupby('name_themes')['parent_id'].sum()
#sets_themes[sets_themes['parent_id'].isna()].groupby('name_themes').agg({'parent_id': })
#sets_themes.sample(100)
sets_themes[sets_themes['parent_id'].isna()].head()
```

```python
#sets[sets['year']==1950]
sets.groupby('year')['name'].nunique().plot(kind = 'bar')
plt.title("Evolución del número de sets por año")
plt.xlabel("Año")
plt.ylabel("Cantidad")
plt.show()
```

```python
parts_year = sets[['year', 'num_parts']].groupby('year', as_index=False).mean()
parts_year.plot(x='year', y='num_parts', color="red")
plt.title("Número promedio de partes por año")
plt.xlabel("Año")
plt.ylabel("Promedio de cantidad de partes")
plt.show()
```
