#1. Trabajo previo del dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import seaborn as sns; sns.set()
```

```python
# Cambiar a lo que corresponda
DATASET_PATH = "dataset"
```

```python
inv_sets = pd.read_csv(DATASET_PATH + "/inventory_sets.csv")
inv_sets.shape
```

```python
inv = pd.read_csv(DATASET_PATH + "/inventories.csv")
inv.head()
```

```python
inv_part = pd.read_csv(DATASET_PATH + "/inventory_parts.csv")
inv_part.head(10)

colors = pd.read_csv(DATASET_PATH + "/colors.csv")
colors.rename(columns = {'id':'color_id'}, inplace = True)

inv_part_colors = inv_part.merge(colors, on='color_id', how='left')
inv_part_colors.shape
inv_part_colors.head()
```

```python
parts = pd.read_csv(DATASET_PATH + "/parts.csv")

parts.shape
#parts[pd.isna(parts['part_cat_id'])]
```

```python
part_cat = pd.read_csv(DATASET_PATH + "/part_categories.csv")
part_cat.rename(columns = {'id':'part_cat_id'}, inplace = True)
part_cat.head()
```

```python
parts_part_cat = parts.merge(part_cat, on='part_cat_id', how='outer', suffixes=('_parts', '_part_cat'))
# Junto parts con parts-categories
parts_part_cat.shape
parts_part_cat.head()
```

```python
sets = pd.read_csv(DATASET_PATH + "/sets.csv")

themes = pd.read_csv(DATASET_PATH + "/themes.csv")
themes.rename(columns = {'id':'theme_id'}, inplace = True)

sets_themes = sets.merge(themes, on='theme_id', how='left', suffixes=('_sets', '_themes'))

sets_themes.shape
sets_themes.head(10)
# Junto sets con themes
```

```python
#sets_themes['set_num'].isna().sum()
```

```python
inv_part_colors_parts_part_cat = inv_part_colors.merge(parts_part_cat, on='part_num', how='left')
inv_part_colors_parts_part_cat.head()
```

```python
inv_sets_sets_themes = inv_sets.merge(sets_themes, on='set_num', how='outer')
inv_sets_sets_themes.shape
inv_sets_sets_themes.head()
```

```python
####################################
inv_parts_inv_sets = inv_part_colors_parts_part_cat.merge(inv_sets_sets_themes, on='inventory_id', how='inner')
inv_parts_inv_sets.shape
#inv_parts_inv_sets.head(10)
```

```python
inv_parts_inv_sets.rename(columns = {'inventory_id':'id'}, inplace = True)
df = inv.merge(inv_parts_inv_sets, on='id', how='inner')
#df.shape
df.head(10)
```
