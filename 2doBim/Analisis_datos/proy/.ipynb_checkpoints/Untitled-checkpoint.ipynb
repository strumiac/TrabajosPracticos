```python
import pandas as pd

df1=pd.DataFrame({
    "id" : ['1','1','2','3','4','5'],
    "fruta" : ['pera','pera2','banana','higo','kiwi','melon']    
})
df2=pd.DataFrame({
    "id_color" : ['1','2','7','4','8'],
    "color" : ['rojo','azul','amarillo','rosa','celeste']    
})
```

```python
#df1.merge(df2,left_on="id",right_on="id_color",how="outer")
df2.rename(columns = {'id_color':'id'}, inplace = True)
df1.merge(df2, on="id", how="outer")

```

```python
df3=pd.DataFrame({
    "id" : ['1','2','7','4','8'],
    "color" : ['rojo','azul','amarillo','rosa','celeste']    
})
df1.merge(df3,on="id",how="outer")
```
