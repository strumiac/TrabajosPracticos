```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import seaborn as sns; sns.set()

DATASET_PATH = "dataset"
sets = pd.read_csv(DATASET_PATH + "/sets.csv")
sets.head(10)
#sets.shape

sets.isna().sum()
```

```python
#sets[sets['year']==1950]
sets.groupby('year')['name'].nunique().plot(kind = 'bar', figsize = (15,5))
plt.title("Evolución del número de sets por año")
plt.xlabel("Año")
plt.ylabel("Cantidad")
plt.show()
```

```python
parts_year = sets[['year', 'num_parts']].groupby('year', as_index=False).mean()
parts_year.plot(x='year', y='num_parts', color="red", figsize = (10,7))
plt.title("Número promedio de partes por año")
plt.xlabel("Año")
plt.ylabel("Promedio de cantidad de partes")
plt.show()
```
