```python
%matplotlib inline
#%matplotlib

import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

# para graficar distribución de características
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib import colors
```

#### Cargo y muestro la imagen de mi elección

```python
#img = cv.imread('peppers.png')
img = cv.imread('pastilla.jpg')

width = 350
height = 150
points = (width, height)
img = cv.resize(img, points, interpolation= cv.INTER_LINEAR)


plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.show()
```

#### 1) Espacio HSV

```python
imgHSV = cv.cvtColor(img, cv.COLOR_BGR2HSV)

plt.imshow(imgHSV)
plt.hsv()
plt.show()
```

#### 2) Espacio BGR y posición (x, y)

```python
imgRGB = cv.cvtColor(img, cv.COLOR_BGR2RGB)

x = np.repeat(np.arange(0,imgRGB.shape[1],1).reshape(1,imgRGB.shape[1]), imgRGB.shape[0], axis = 0)
y = np.repeat(np.arange(0,imgRGB.shape[0],1).reshape(imgRGB.shape[0],1), imgRGB.shape[1], axis = 1)
x = np.reshape(x, (x.shape[0],x.shape[1],1))
y = np.reshape(y, (y.shape[0],y.shape[1],1))

imgRGB_xy = np.concatenate((imgRGB/255, y/y.max(), x/x.max()), axis = 2)
imgRGB_xy.shape

plt.imshow(imgRGB)
plt.show()
```

#### 3)i) Grafico la distribución de características para HSV

```python
H, S, V = cv.split(imgHSV)
pixel_colors = imgHSV.reshape((np.shape(imgHSV)[0]*np.shape(imgHSV)[1], 3))
norm = colors.Normalize(vmin=-1.,vmax=1.)
norm.autoscale(pixel_colors)
pixel_colors = norm(pixel_colors).tolist()

fig = plt.figure(figsize = (8,8))
axis = fig.add_subplot(1, 1, 1, projection="3d")
axis.scatter(H.flatten(), S.flatten(), V.flatten(), marker = ".", facecolors = pixel_colors)
axis.set_xlabel("H", fontsize = 10)
axis.set_ylabel("S", fontsize = 10)
axis.set_zlabel("V", fontsize = 10)
axis.set_title("Distribución HSV", fontsize = 12)
plt.show()
```

#### 3)ii) Grafico la distribución de características para RGB

```python
R, G, B = cv.split(imgRGB)
pixel_colors = imgRGB.reshape((np.shape(imgRGB)[0]*np.shape(imgRGB)[1], 3))
norm = colors.Normalize(vmin=-1.,vmax=1.)
norm.autoscale(pixel_colors)
pixel_colors = norm(pixel_colors).tolist()

fig = plt.figure(figsize = (8,8))
axis = fig.add_subplot(1, 1, 1, projection="3d")
axis.scatter(R.flatten(), G.flatten(), B.flatten(), marker = ".", facecolors = pixel_colors)
axis.set_xlabel("R", fontsize = 10)
axis.set_ylabel("G", fontsize = 10)
axis.set_zlabel("B", fontsize = 10)
axis.set_title("Distribución RGB", fontsize = 12)
plt.show()
```

#### 4)i) Imagen posterizada y máscaras para HSV

```python
features = imgHSV.reshape((-1,3))
print('Dimensión de la matriz de características (espacio HSV): {}'.format(features.shape))
features = np.float32(features)
```

```python
criteria = (cv.TERM_CRITERIA_MAX_ITER + cv.TERM_CRITERIA_EPS, 10, 1.0)
K = 3

compact, label, color_clusters = cv.kmeans(features, K, None, criteria, 10, cv.KMEANS_RANDOM_CENTERS)

color_clusters = np.uint8(color_clusters)
print(f'Centroides: \n {color_clusters} \n')

compact = np.uint8(compact)
print('Compactación: \n {} \n'.format(compact))
label[:4]
```

```python
img_out = color_clusters[label.flatten()]
img_out = img_out.reshape((imgHSV.shape))

fig = plt.figure(figsize = (6,6))
plt.imshow(img_out)
plt.title("Imagen posterizada para HSV", fontsize = 12)
plt.hsv()
plt.show()
```

```python
mask_list = list()
for cluster in color_clusters:
    mask = np.zeros(img_out.shape[:2],'uint8')
    mask[np.where((img_out == cluster).all(axis=2))] = 255
    mask_list.append(mask)

for i, mask in enumerate(mask_list):
    img_segmented = cv.bitwise_and(imgHSV, imgHSV, mask = mask)
    fig = plt.figure(figsize = (6,6))
    plt.imshow(mask, cmap = 'gray')
    plt.title(f"mascara {i+1}", fontsize = 12)
    plt.show()
```

#### 4)ii) Imagen posterizada y máscaras para RGB

```python
features = imgRGB_xy.reshape((-1,5))
print('Dimensión de la matriz de características (espacio Gray XY): {}'.format(features.shape))
features = np.float32(features)
features
```

```python
# Criterio de detención (Epsilon e iteraciones máximas) y número de clústers (K)
criteria = (cv.TERM_CRITERIA_MAX_ITER + cv.TERM_CRITERIA_EPS, 10, 1.0)
K = 5

compact, label, color_clusters = cv.kmeans(features, K, None, criteria, 10, cv.KMEANS_RANDOM_CENTERS)

norma = np.array([255, 255, 255, y.max(), x.max()])
color_clusters = np.uint(color_clusters * norma)
#color_clusters = np.uint(color_clusters)
print(f'Centroides: \n {color_clusters} \n')

compact = np.uint8(compact)
print('Compactación: \n {} \n'.format(compact))
label[:4]
```

```python
img_out = color_clusters[label.flatten()]
img_out = img_out.reshape(imgRGB_xy.shape)

fig = plt.figure(figsize = (6,6))
plt.title("Imagen posterizada para RGB x-y", fontsize = 12)
plt.imshow(img_out[:,:,0:3])
plt.show()
```

```python
mask_list = list()
for cluster in color_clusters:
    mask = np.zeros(img_out.shape[:2],'uint8')
    mask[np.where((img_out == cluster).all(axis=2))] = 255
    mask_list.append(mask)

for i, mask in enumerate(mask_list):
    img_segmented = cv.bitwise_and(imgRGB, imgRGB, mask = mask)
    fig = plt.figure(figsize = (6,6))
    plt.imshow(mask, cmap = 'gray')
    plt.title(f"mascara {i+1}", fontsize = 12)
    plt.show()
```
