```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
tito1 = cv.imread('fig/WP_B.png')

tito1RGB = cv.cvtColor(tito1,cv.COLOR_BGR2RGB)

fig = plt.figure()
ax1 = plt.subplot(121)
ax1.imshow(tito1RGB)
ax1.set_title('Tito 1 original')

tito1RGB_wpch = np.round(tito1RGB/np.max(tito1RGB, axis=(0,1)))
ax2 = plt.subplot(122)
ax2.imshow(tito1RGB_wpch.round())
ax2.set_title('Tito 1 White patch')


plt.show()
```
