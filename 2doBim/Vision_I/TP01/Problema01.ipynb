```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
tito1 = cv.imread('fig/CoordCrom_1.png')

tito1RGB = cv.cvtColor(tito1,cv.COLOR_BGR2RGB)

fig = plt.figure()
ax1 = plt.subplot(121)
ax1.imshow(tito1RGB)
ax1.set_title('Tito 1 original')

sum_comp=tito1RGB.sum(axis=2)[:,:,None] #agrego None para que sea una matriz 3D
sum_comp[sum_comp == 0] = 1  #elimino los ceros para poder dividir
tito1RGB_crom = ((tito1RGB/sum_comp)*255).astype(int)
ax2 = plt.subplot(122)
ax2.imshow(tito1RGB_crom)
ax2.set_title('Tito 1 cromatico')

plt.show()
```
