```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
image = cv.imread('fig/CoordCrom_1.png')

B, G, R = cv.split(image)
Bint = B.astype(int)
Gint = G.astype(int)
Rint = R.astype(int)

sum_rgb = Rint + Gint + Bint
sum_rgb[sum_rgb == 0] = 1

fig = plt.figure()
ax1 = plt.subplot(121)
ax1.imshow(cv.merge([R, G, B]))
ax1.set_title('Tito 1 original')
ax2 = plt.subplot(122)
ax2.imshow(cv.merge([Rint/sum_rgb, Gint/sum_rgb, Bint/sum_rgb]))
ax2.set_title('Tito 1 coord cromtica')
plt.show()
```
