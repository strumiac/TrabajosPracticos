```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
image = cv.imread('fig/WP_B.png')
#image = cv.imread('fig/test_blue.png')

B, G, R = cv.split(image)

maxb = np.amax(B)
maxg = np.amax(G)
maxr = np.amax(R)
print ('maxb', maxb,'maxg', maxg, 'maxr', maxr)
sb = 255/maxb
sg = 255/maxg
sr = 255/maxr


fig = plt.figure()
ax1 = plt.subplot(121)
ax1.imshow(cv.merge([R, G, B]))
ax1.set_title('Tito B original')
ax2 = plt.subplot(122)
ax2.imshow(cv.merge([(sr*R).astype('uint8'), (sg*G).astype('uint8'), (sb*B).astype('uint8')]))
ax2.set_title('Tito B white patch')
plt.show()
```
