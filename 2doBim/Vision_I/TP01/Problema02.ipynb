```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
tito1 = cv.imread('fig/WP_B.png')

tito1RGB = cv.cvtColor(tito1,cv.COLOR_BGR2RGB)

fig = plt.figure()
ax1 = plt.subplot(121)
ax1.imshow(tito1RGB)
ax1.set_title('Tito 1 original')

#x=np.max(tito1RGB, axis=(0,1))
#x[]

tito1RGB_wpch = ((tito1RGB/1)*255).astype(int)
ax2 = plt.subplot(122)
ax2.imshow(tito1RGB_wpch)
ax2.set_title('Tito 1 White patch')


plt.show()

np.max(tito1RGB, axis=(0,1)).shape
print(tito1RGB)

#print(np.max(tito1RGB, axis=(0,1)))

#print((tito1RGB/np.max(tito1RGB, axis=(0,1)))*255)
```

```python
tito2 = cv.imread('fig/WP_O.png')

tito2RGB = cv.cvtColor(tito2,cv.COLOR_BGR2RGB)

fig = plt.figure()
ax1 = plt.subplot(121)
ax1.imshow(tito2RGB)
ax1.set_title('Tito 2 original')

tito2RGB_wpch = ((tito2RGB/np.max(tito2RGB, axis=(0,1)))*255).astype(int)
ax2 = plt.subplot(122)
ax2.imshow(tito2RGB_wpch.round())
ax2.set_title('Tito 2 White patch')


plt.show()
```

```python
tito3 = cv.imread('fig/WP_R.png')

tito3RGB = cv.cvtColor(tito3,cv.COLOR_BGR2RGB)

fig = plt.figure()
ax1 = plt.subplot(121)
ax1.imshow(tito3RGB)
ax1.set_title('Tito 3 original')

tito3RGB_wpch = ((tito3RGB/np.max(tito3RGB, axis=(0,1), keepdims=True))*255).astype(int)
ax2 = plt.subplot(122)
ax2.imshow(tito3RGB_wpch.round())
ax2.set_title('Tito 3 White patch')

plt.show()
(np.max(tito3RGB, axis=(0,1), keepdims=True))
```
