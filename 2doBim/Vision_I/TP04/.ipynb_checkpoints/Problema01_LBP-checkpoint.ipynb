```python
%matplotlib inline

import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import glob

power_2 = np.array([128, 64, 32, 16, 8, 4, 2, 1])
w_hist = {}

for image in glob.glob('*.png'):
    img = cv.imread(image)
    img_RGB = cv.cvtColor(img,cv.COLOR_BGR2RGB)
    gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    gray_pad = np.pad(gray, pad_width = 1, mode = 'edge')
    #print ("La dimension de la imagen es", gray.shape,"y la dimensin con pading es",gray_pad.shape)
    gray_LBP = np.zeros_like(gray)
    for i in range(1, gray_pad.shape[0]-1):
        for j in range(1, gray_pad.shape[1]-1):
            ventana = (np.append(np.append(np.append(gray_pad[i-1,j-1:j+2],gray_pad[i,j+1]),np.flip(gray_pad[i+1,j-1:j+2])),gray_pad[i,j-1]))
            gray_LBP[i-1,j-1] = np.sum((ventana >= gray_pad[i,j])*power_2)
    
    #comp.append(np.ravel(gray_LRB))
    #w_hist[image[0:7]]=np.ravel(gray_LBP)
    
    fig, ax = plt.subplots(1, 3)
    fig.set_size_inches(15,4)
    ax[0].imshow(img_RGB)
    ax[0].set_title('Archivo {}'.format(image))
    ax[1].imshow(gray_LBP, cmap = 'gray')
    ax[1].set_title('Textura LBP {}'.format(image))
    ax[2].hist(np.ravel(gray_LBP), bins = 256)
    ax[2].set_title('Histograma LBP de {}'.format(image))
    plt.show()
    
    
    #hist = cv.calcHist(gray_LBP, [0, 1, 2], None, [8, 8, 8], [0, 255, 0, 255, 0, 255])#[0, 1, 2]  [8, 8, 8]
    hist = cv.calcHist(gray_LBP, [0], None, [256], [0, 255])
    hist = cv.normalize(hist, hist).flatten()
    w_hist[image[0:7]] = hist
```
