### Importación de librerías

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import VarianceThreshold

# Entrenamiento de modelos de prueba
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.base import clone

# Evaluación de modelos de prueba
from sklearn.metrics import roc_auc_score
from sklearn.metrics import mean_squared_error
from sklearn import metrics


DATASET_PATH = "dataset"
```

### Merge de los distintos .csv

```python
sets = pd.read_csv(DATASET_PATH + "/sets.csv")
themes = pd.read_csv(DATASET_PATH + "/themes.csv")

themes.rename(columns = {'id':'theme_id'}, inplace = True)
sets = sets.merge(themes, on='theme_id', how='left')
sets.rename(columns = {'name_x':'set_name','name_y':'theme_name'}, inplace = True)

inv = pd.read_csv(DATASET_PATH + "/inventories.csv")
inv = inv.merge(sets, on='set_num', how='left')

inv_part = pd.read_csv(DATASET_PATH + "/inventory_parts.csv")
colors = pd.read_csv(DATASET_PATH + "/colors.csv")

colors.rename(columns = {'id':'color_id'}, inplace = True)

inv_part = inv_part.merge(colors, on='color_id', how='left')
inv_part.rename(columns = {'inventory_id':'id'}, inplace = True)

df = inv_part.merge(inv, on='id', how='left')
print ('Cantidad de colores: ', df.rgb.nunique())
print ('Cantidad de sets: ', df.set_num.nunique())
print ('Cantidad de temáticas: ',df.theme_name.nunique())
```

### Cardinalidad de las clases

```python
# Trabajo para que quede en 7 clases agrupadas

df_g['theme_group'] = 0
group = range(0, 80, 10)
for i in range(7):
    for j in range(df_g.shape[0]):
        if (df_g.iat[j,1] > group[i] and df_g.iat[j,1] <= group[i + 1]):
            df_g.iloc[j,2] = 'Clase{}'.format(i+1)
# Me queda una clase con 74 sin grupo, lo sumo al grupo 7
df_g.iloc[df_g.rgb == 74, 2] = 'Clase7'
df_g = df_g.drop(['rgb'], axis = 'columns')
df_g
```

```python
# Frecuency encoder RGB
rgb_frecuency = df.rgb.value_counts(normalize = True).rename_axis('rgb').reset_index(name='rgb_frec')

# Frecuency encoder temáticas
theme_frecuency = df.theme_name.value_counts(normalize = True).rename_axis('theme_name').reset_index(name='theme_frec')

#theme_frecuency.theme_frec.unique().shape
#rgb_frecuency.head(10)
theme_frecuency
```

### Codificación para los 75 colores mas relevantes

```python
# Se agregan n columnas para caracterizar cada theme por sus n colores mas relevantes
n = 75

for i in range(n):
    df['rgb{}'.format(i+1)]=0
```

```python
df_15 = df.copy()
df_15 = df_15.merge(rgb_frecuency, on='rgb', how='left')
df_15 = df_15.merge(theme_frecuency, on='theme_name', how='left')
print(df_15.shape)
df_15.head()
```

```python
#for tematica in df.theme_name.unique():
for clase in df_15.theme_name.unique():
    #df_rgb = df.loc[df.theme_name == tematica].rgb.value_counts()[0:n].rename_axis('rgb').reset_index(name='sum')
    df_15_rgb = df_15.loc[df_15.theme_name == clase].rgb.value_counts()[0:n].rename_axis('rgb').reset_index(name='sum')
    df_15_rgb = df_15_rgb.merge(rgb_frecuency, on='rgb', how='left').sort_values('rgb_frec', ascending = False)
    if (n > df_15_rgb.shape[0]):
        for i in range(df_15_rgb.shape[0]):
            #df.loc[df.theme_name == tematica, 'rgb{}'.format(i+1)] = df_rgb.iat[i,2]
            df_15.loc[df_15.theme_name == clase, 'rgb{}'.format(i+1)] = df_15_rgb.iat[i,2]
    else:
        for i in range(n):
            #df.loc[df.theme_name == tematica, 'rgb{}'.format(i+1)] = df_rgb.iat[i,2]
            df_15.loc[df_15.theme_name == clase, 'rgb{}'.format(i+1)] = df_15_rgb.iat[i,2]
```

```python
df_15 = df_15.drop(['rgb', 'set_num', 'num_parts', 'theme_name', 'rgb_frec', 'theme_frec'], axis = 'columns')
df_15 = df_15.drop_duplicates(ignore_index = True)
df_15.head()
```

### Separación del dataset en train y test (0.3)

```python
X = df_15d[['rgb1' ,'rgb2' ,'rgb3' ,'rgb4' ,'rgb5' ,'rgb6','rgb7','rgb8','rgb9','rgb10',
          'rgb11','rgb12','rgb13','rgb14','rgb15','rgb16','rgb17','rgb18','rgb19','rgb20',
          'rgb21','rgb22','rgb23','rgb24','rgb25','rgb26','rgb27','rgb28','rgb29','rgb30',
          'rgb31','rgb32','rgb33','rgb34','rgb35','rgb36','rgb37','rgb38','rgb39','rgb40',
          'rgb41','rgb42','rgb43','rgb44','rgb45','rgb46','rgb47','rgb48','rgb49','rgb50',
          'rgb51','rgb52','rgb53','rgb54','rgb55','rgb56','rgb57','rgb58','rgb59','rgb60',
          'rgb61','rgb62','rgb63','rgb64','rgb65','rgb66','rgb67','rgb68','rgb69','rgb70',
          'rgb71','rgb72','rgb73','rgb74','rgb75']]
y = df_15d[['theme_group']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train.sample(10)
```

### Reducción de dimesiones

Primero buscamos variables de varianza 0 (constante).

```python
var = np.std(X_train, axis=0)**2
var[var == 0]
```

```python
from sklearn.feature_selection import VarianceThreshold
sel_0var = VarianceThreshold(threshold=0)
sel_0var.fit(X_train);
X_train_red = sel_0var.transform(X_train)
X_test_red = sel_0var.transform(X_test)
print(sel_0var.variances_)
X_train_red.shape, X_test_red.shape
```

Buscamos variables de baja varianza.

```python
sel = VarianceThreshold(threshold=1.0e-04)
sel.fit(X_train_red);
X_train_red2 = sel.transform(X_train_red)
X_test_red2 = sel.transform(X_test_red)
print(sel.variances_)
X_train_red2.shape, X_test_red2.shape
```

```python
df_train = pd.DataFrame(X_train_red2)
corr = df_train.corr()
corr.shape
#corr = X_train_red.corr()
corr
```

```python
sns.set(font_scale=0.8, rc = {'figure.figsize':(10,10)})
sns.heatmap(corr, cmap = sns.diverging_palette(220,10,as_cmap = True), annot = True, fmt = ".2f")
```

### Entrenamiento del modelo

De las 7 clases elegimos las primeras 2 para entrenar.

#### Análisis para la Clase 0

```python
X_train = X_train_red2
y_train = y_train_encoded[:,0]
X_test = X_test_red2
y_test = y_test_encoded[:,0]

models = [("KNN", KNeighborsClassifier(n_neighbors = 7)),
          ("LogisticRegression", LogisticRegression(random_state = 42, C = 1000, solver = 'lbfgs'))]

results = {}
for m in models:
    model_name = m[0]
    results[model_name] = {}
    model = clone(m[1])
      
    model.fit(X_train, y_train)
    pred_train = model.predict_proba(X_train)
    pred_test = model.predict_proba(X_test)
    
    results[model_name]["train"] = roc_auc_score(y_train, pred_train[:,1])
    results[model_name]["test"] =non_red_score_test = roc_auc_score(y_test, pred_test[:,1])

df_results = pd.DataFrame(results).T


clase0_knn = KNeighborsClassifier(n_neighbors = 7)
clase0_knn.fit(X_train, y_train)
y_pred_knn = clase0_knn.predict(X_test)

print("El valor de accuracy para KNN es: {}".format(round(metrics.accuracy_score(y_test, y_pred_knn),2)))
print("El valor de precision para KNN es: {}".format(round(metrics.precision_score(y_test, y_pred_knn),2)))
print("El valor de recall para KNN es: {}".format(round(metrics.recall_score(y_test, y_pred_knn),2)))

clase0_log = LogisticRegression(random_state = 42, C = 1000, solver = 'lbfgs')
clase0_log.fit(X_train, y_train)
y_pred_log = clase0_log.predict(X_test)
print("El valor de accuracy para Reg Log es: {}".format(round(metrics.accuracy_score(y_test, y_pred_log),2)))
print("El valor de precision para Reg Log es: {}".format(round(metrics.precision_score(y_test, y_pred_log),2)))
print("El valor de recall para Reg Log es: {}".format(round(metrics.recall_score(y_test, y_pred_log),2)))

print('\nComparativa ROC para los dos modelos empleados Clase 0:')
df_results
```
