## Clasificación por sexo

```python
#Importo librerias clásicas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importo librerias sklearn

from sklearn import metrics
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

RANDOM_STATE = 12
```

```python
#Leo el csv

df = pd.read_csv("possum.csv")
df.head()
```

```python
# Chequeo posibles NaN

df[pd.isnull(df).any(axis=1)]
```

```python
# Transformación de variables categóricas e imputación de NaN

df.sex = df.sex.map({'m': 0, 'f': 1})

df.age.fillna(df.age.mean(), inplace=True)
df.footlgth.fillna(df.footlgth.mean(), inplace=True)

df
```

```python
# Borro algunas columnas

df.drop(['case', 'site', 'Pop'], axis=1, inplace=True)
```

```python
# Divido en train y test

X_train, X_test, y_train, y_test = train_test_split(df.drop(['sex'], axis=1), df.sex.copy(), test_size = 0.3, random_state = RANDOM_STATE)
X_train.shape
```

#### Entrenamiento con regresión logística

```python
# Entrenando regresión logistica

sex_model = LogisticRegression()
sex_model.fit(X_train, y_train)
```

```python
y_pred = pd.Series(sex_model.predict(X_test))
y_test = y_test.reset_index(drop = True)
y_comp = pd.concat([y_test, y_pred], axis = 1)
y_comp.columns = ['True', 'Prediction']
y_comp.head()
```

```python
# desempeño del modelo

print("El valor de accuracy es: {}".format(round(metrics.accuracy_score(y_test, y_pred),2)))
print("El valor de precision es: {}".format(round(metrics.precision_score(y_test, y_pred),2)))
print("El valor de recall es: {}".format(round(metrics.recall_score(y_test, y_pred),2)))
```

```python
# para visualizar dibujo la matriz de confusión

conf_matriz = metrics.confusion_matrix(y_test, y_pred)

labels = [0, 1]
fig, ax = plt.subplots()
tick_marks = np.arange(len(labels))
plt.xticks(tick_marks, labels)
plt.yticks(tick_marks, labels)

sns.heatmap(pd.DataFrame(conf_matriz), annot=True, cmap="PuBu", fmt='g')
ax.xaxis.set_label_position("bottom")
plt.title('Matriz de confusión')
plt.ylabel('True')
plt.xlabel('Predicción')
plt.show()
```

#### Entrenamiento con KNN

```python
sex_knn = KNeighborsClassifier(n_neighbors = 5)
sex_knn.fit(X_train, y_train)
```

```python
y_pred_knn = pd.Series(sex_knn.predict(X_test))
y_test = y_test.reset_index(drop = True)
y_comp_knn = pd.concat([y_test, y_pred_knn], axis = 1)
y_comp_knn.columns = ['True', 'Prediction']
y_comp_knn.head()
```

```python
# desempeño del modelo

print("El valor de accuracy es: {}".format(round(metrics.accuracy_score(y_test, y_pred_knn),2)))
print("El valor de precision es: {}".format(round(metrics.precision_score(y_test, y_pred_knn),2)))
print("El valor de recall es: {}".format(round(metrics.recall_score(y_test, y_pred_knn),2)))
```

```python
# para visualizar dibujo la matriz de confusión

conf_matriz = metrics.confusion_matrix(y_test, y_pred_knn)

labels = [0, 1]
fig, ax = plt.subplots()
tick_marks = np.arange(len(labels))
plt.xticks(tick_marks, labels)
plt.yticks(tick_marks, labels)

sns.heatmap(pd.DataFrame(conf_matriz), annot=True, cmap="PuBu", fmt='g')
ax.xaxis.set_label_position("bottom")
plt.title('Matriz de confusión para KNN')
plt.ylabel('True')
plt.xlabel('Predicción')
plt.show()
```

```python
# Divido en train y test

X_train, X_test, y_train, y_test = train_test_split(df.drop(['totlngth'], axis=1), df.totlngth.copy(), test_size = 0.3, random_state = RANDOM_STATE)
X_train.shape
```

```python
#grafico la matriz de correlacion para X_train. No se observa mucha correlación entre features

plt.figure(figsize = (12,8))
sns.heatmap(X_train.corr(), annot = True, cmap = 'RdPu')
```

```python
reg_totlngth = LinearRegression()
reg_totlngth.fit(X_train,y_train)
y_pred = reg_totlngth.predict(X_test)

# Coeficientes
print("b", reg_totlngth.intercept_)
print(reg_totlngth.coef_)
```

```python
# desempeño del modelo

print("El valor de MAE es: {}".format(round(metrics.mean_absolute_error(y_test, y_pred),2)))
print("El valor de MSE es: {}".format(round(metrics.mean_squared_error(y_test, y_pred),2)))
print("El valor de RMSE es: {}".format(round(np.sqrt(metrics.mean_squared_error(y_test, y_pred)),2)))
```

```python
fig, axs = plt.subplots()
axs.scatter(y_test, y_pred)
axs.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'g', lw = 3)
axs.set_xlabel('Medido')
axs.set_ylabel('Predicción')
plt.show()
```
